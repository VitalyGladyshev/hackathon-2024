```python
chapters_list_corr = []

for chpt in chapters_list[:-1]:
    if chpt != "ОСТРОВ ХАЙНАНЬ":
        chapters_list_corr.append(chpt)
    else:
        chapters_list_corr.append("ХАЙНАНЬ")

chapters_list_corr
```

```python
for chapter_name in chapters_list_corr:
    pos_list = [m.start() for m in re.finditer(chapter_name, chapters[chapter_name])]
    print(f"{chapter_name} pos list: {pos_list}")
    if len(pos_list) > 1:
        chapters[chapter_name] = chapters[chapter_name][pos_list[1]+1+len(chapter_name):]
    else:
        chapters[chapter_name] = chapters[chapter_name][1+len(chapter_name):]
```

```python
for chapter_name in chapters_list_corr:
    pos_list = [m.start() for m in re.finditer(chapter_name, chapters[chapter_name])]
    print(f"{chapter_name} pos list: {pos_list}")
    pos_list.reverse()  # Удаляем в обратном порядке
    for pos in pos_list:
        cur_len = len(chapter_name)
        cur_pos = pos + cur_len
        while(not chapters[chapter_name][cur_pos].isalpha()):
            cur_len += 1
            cur_pos += 1
        print(chapters[chapter_name][pos:pos+cur_len])
        chapters[chapter_name] = chapters[chapter_name][:pos] + chapters[chapter_name][pos+cur_len:]
```

```python
sub_chpt_names = defaultdict(list)
next_ind = 1
sub_chpt_names[chapters_list[0]] = []
chapters_list_len = len(chapters_list)

for line in lines[1:]:
    if next_ind < chapters_list_len and line == chapters_list[next_ind]:
        print(f"line: {line} ind: {next_ind} chapters_list[next_ind]: {chapters_list[next_ind]}")
        next_ind += 1
        if next_ind == len(chapters) + 1:
            break
    else:
        print(f"\tnext_ind-1: {next_ind-1} chapters_list[next_ind-1]: {chapters_list[next_ind-1]} line: {line}")
        if chapters_list[next_ind-1] != "ОСТРОВ ХАЙНАНЬ":
            sub_chpt_name = ""
            if line != "Г еография и климат":
                sub_chpt_name = line.upper().strip()
            else:
                sub_chpt_name = "ГЕОГРАФИЯ И КЛИМАТ"

            find_test = chapters[chapters_list[next_ind-1]].find(sub_chpt_name)
            if find_test >= 0:
                sub_chpt_names[chapters_list[next_ind-1]].append(sub_chpt_name)
            else:
                print(f"!{sub_chpt_name} не найдено!")
        else:
            sub_chpt_names["ХАЙНАНЬ"].append(line.upper().strip())
```

```python
subchapters = defaultdict(dict)

for chpt in headers:
    print(f"{chpt} len: {len(chapters[chpt])}")
    if len(sub_chpt_names[chpt]):
        # print(len(sub_chpt_names[chpt]))
        for i in range(len(sub_chpt_names[chpt])):
            sub_chpt_name = sub_chpt_names[chpt][i]

            start = chapters[chpt].find(sub_chpt_name) + len(sub_chpt_name) + 1
            end = 0
            # print(f"\t\ti: {i} len(sub_chpt_names[chpt]): {len(sub_chpt_names[chpt])}")
            if i != len(sub_chpt_names[chpt]) - 1:
                sub_chpt_name_next = sub_chpt_names[chpt][i+1]
                end = chapters[chpt].find(sub_chpt_name_next)
                if end < 0:
                    print("!!!" + sub_chpt_name_next)
                subchapters[chpt][sub_chpt_name] = chapters[chpt][start:end]
            else:
                end = -1
                subchapters[chpt][sub_chpt_name] = chapters[chpt][start:]

            if end > 0:
                print(f"\t{sub_chpt_name} [{start}:{end}] len: {end - start}")
            else:
                print(f"\t{sub_chpt_name} [{start}:{end}] len: {len(chapters[chpt]) - start}")

    else:
        subchapters[chpt][chpt] = chapters[chpt]
```
